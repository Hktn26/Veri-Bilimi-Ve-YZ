```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn import model_selection
import matplotlib.pyplot as plt
from sklearn.linear_model import RidgeCV
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

df = pd.read_csv('Hitters.csv')
```

```python
# Eksik verileri cikarma
df = df.dropna()
df.isnull().sum()
```

```python
dms = pd.get_dummies(df[["League", "Division", "NewLeague"]]) 
# Kategorik değişkenleri one-hot encoding ile dönüştür
y = df["Salary"] 
# Hedef değişken
X_ = df.drop(["Salary", "League", "Division", "NewLeague"], axis=1).astype("float64") 
# Bu satırda, Salary ve kategorik sütunlar (çünkü onların one-hot versiyonunu az sonra ekleyeceğiz) çıkarılır.
#Kalan sayısal sütunlar float64 veri tipine dönüştürülür (bazı algoritmalar veri tipi konusunda hassas olabilir).
X = pd.concat([X_, dms[["League_N", "Division_W", "NewLeague_N"]]], axis=1) 
# Özellikleri birleştir


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=26) 
# Veriyi eğitim ve test setlerine ayır
```

```python
alphas = np.logspace(-2, 4, 1000)
alphas1 = 10**np.linspace(10, -2, 100)*0.5
```

```python
model = Ridge().fit(X_train, y_train)
# Ridge modelini eğitim seti ile eğit
```

```python
model = Ridge().fit(X_train, y_train) # Ridge regresyon modelini eğit
y_pred = model.predict(X_test) # Test seti üzerinde tahmin yap
np.sqrt(mean_squared_error(y_test, y_pred)) # Test seti RMSE değerini hesapla
```

```python
y_pred = model.predict(X_test)
# Test seti üzerinde tahmin yap
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
# RMSE hesapla
print("RMSE: ", rmse)
# RMSE değerini yazdır
r2 = r2_score(y_test, y_pred)
# R^2 hesapla
print("R^2: ", r2)
# R^2 değerini yazdır
```

```python
## MODELING + TUNING
ridgecv = make_pipeline(
    StandardScaler(),
    RidgeCV(alphas=alphas, scoring="neg_mean_squared_error", cv=10)
)
ridgecv.fit(X_train, y_train)
```

```python
best_alpha = ridgecv.named_steps['ridgecv'].alpha_
print("En iyi alpha:", best_alpha)
```

```python
#! Model performansını değerl

y_pred = ridgecv.predict(X_test)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("RMSE:", rmse)
r2 = r2_score(y_test, y_pred)
print("R2:", r2)
```
